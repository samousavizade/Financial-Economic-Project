```python
import numpy as np
import pandas as pd
#from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import yfinance as yf
```

# Top 5 Coins by Market Cap

```python
data = pd.DataFrame()
# define the ticker symbol
symbols = ["BTC-USD", 
           "ETH-USD",
           "BNB-USD",
           "ADA-USD",
           "XRP-USD",
           ]

for symbol in symbols:
    #get data on this ticker
    tickerData = yf.Ticker(symbol)
    #get the historical prices for this ticker
    symbol_close_data = tickerData.history(period="1d", start="2021-06-01", end="2022-06-01")["Close"]
    data[symbol] = symbol_close_data

data
```

```python
data.to_csv("crypto_data_3.csv")
```
